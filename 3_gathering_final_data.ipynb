# Last data cleaning!

In this notebook, we will combine out two datasets and do the final cleaning. 

![Corgi](https://media.giphy.com/media/OqBvKFGztLpni/giphy.gif)

```python
import pandas as pd
import numpy as np

pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
df1 = pd.read_csv('../7_project/data/dog_show/csv_files/cleaned_messy_dogs.csv')
df1.head()
```

```python
df2 = pd.read_csv('../7_project/data/dog_scraped/csv_files/dog_data_scraped.csv')
df2.head()
```

## Let's join the dataframes

```python
# Joining the dataframes                       
df3 = pd.merge(df2, df1, how ='inner', on ='breed')
df3.head()
```

```python
print(f'The length of the new dataframe is {len(df3)}, and it has {len(df3.columns)} columns:, {df3.columns}')
```

```python
df3.duplicated().sum()
```

Let's remove the columns that are the same or similar from the two different dataset after further inspection

```python
drop_cols = ['avg_pur_USD', 'intelligence_rank', 'pop_us']
```

```python
df3 = df3.drop(drop_cols, axis=1)
```

```python
df3.head()
```

Now it's ready for analysis 

```python
df3 = df3.reset_index(drop=True).copy()
```

```python
df3.to_csv('../7_project/data/df_ready/csv_files/final_clean_data.csv', index=False)
```
